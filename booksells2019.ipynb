```python
import plotly.express as px

rating_by_genre = df.groupby(['Year', 'Genre'], as_index = False) [['User Rating']].mean().round(2)

genre_rating = px.bar(rating_by_genre, title = 'Average User Rating per Year by Genre (Numbers Rounded to the Nearest 100th)', x='Year', y = 'User Rating', color = 'Genre', text_auto=True, barmode = 'group')

genre_rating.show()
```

```python
rating_price_data = df.groupby(['Price'], as_index = False) [['User Rating']].mean()

price_by_rating = px.scatter(df, title = 'Rating in relation to price', x = 'User Rating', y = 'Price')

price_by_rating.show()
```

```python
avg_price_data = df.groupby(['Year'], as_index = False) [['Price']].mean()
avg_price_overtime = px.line(avg_price_data, title = 'Average Price from 2009-2019', x = 'Year', y = 'Price')
avg_price_overtime.show()
```

```python
author_top = df['Author'].value_counts().sort_values(ascending = False).head(15).index
author_data = df[df['Author'].isin(author_top)]
author_domination = px.bar(author_data, title = 'Top 15 Authors with Most Appearences on Top 50 List', color = 'Author', y = 'Author', category_orders={'Author': author_top})
author_domination.show()
```
